```python
import pandas as pd
import numpy as np
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,log_loss,confusion_matrix,recall_score,precision_score
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import time
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB,BernoulliNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV,KFold
import cv2
import pickle
```

```python
data=pd.read_csv('./MNIST_DATASET/train.csv',sep=',')
```

```python
data
```

```python
# # Remove data having label as 0
# data.drop(data[data.label==0].index,axis=0,inplace=True)
```

```python
# change every pixel value to 0 where label is 0
data.loc[data.label==0,data.columns!='label']=0
```

```python
# save this data frame
data.to_csv('./MNIST_DATASET/zero_as_blankSpace_train.csv')
```

```python
X=data.iloc[:,1:]
y=data.iloc[:,0]
```

```python
print(X.shape,y.shape)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,shuffle=True,random_state=1911)
```

```python
print(X_train.shape,y_train.shape,X_test.shape,y_test.shape)
```

# SVM

```python
model=SVC(probability=True)
t=time.time()
model.fit(X_train,y_train)
t=time.time()-t

p=time.time()
y_pred=model.predict(X_test)
p=time.time()-p


print('accuracy_score',round(accuracy_score(y_test,y_pred),3))
print('Trainin time : ',t)
print('Prediction time : ',p)

y_proba=model.predict_proba(X_test)
print('logg loss : ',round(log_loss(y_test,y_proba),3))
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(precision_score(y_test,y_pred,average='micro'))
print(recall_score(y_test,y_pred,average='micro'))
```
